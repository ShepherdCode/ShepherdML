```python
# Image Segmentation

filename="fiddler.jpeg"
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread(filename)
imgplot = plt.imshow(img)
plt.show()
```

```python
img.shape
# Image has 3 colors/pixel over 821 * 1200 pixels
```

```python
RGB = img.reshape(-1,3)
RGB.shape
```

```python
from sklearn.cluster import KMeans
K = 8  # 
km = KMeans(n_clusters=K)
km.fit(RGB)
projection=km.cluster_centers_[km.labels_]
```

```python
revised=projection.reshape(img.shape)
revised=(revised)/255  # convert RGB integer to float(0..1)
revised=(revised+0.1)%1  # rotate color wheel by arbitrary amount
imgplot = plt.imshow(revised)
plt.show()
# We successfully clustered face|shirt|fiddle|wall|window
# but incomplete separation of windowsill|arm or fiddle|hair.
```
