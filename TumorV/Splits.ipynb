```python
class all_splits():
    def load_csv(self,FULLPATH):
        self.X_train_sets=[]
        self.y_train_sets=[]
        self.X_valid_sets=[]
        self.y_valid_sets=[]
        df = pd.read_csv(FULLPATH)
        self.FOLDS=5
        for i in range(self.FOLDS):
            Xt='f'+str(i)+'_TRAIN_X'
            yt='f'+str(i)+'_TRAIN_y'
            Xv='f'+str(i)+'_TEST_X'
            yv='f'+str(i)+'_TEST_y'
            X_train=df[Xt].dropna().tolist()
            y_train=df[yt].dropna().tolist()
            X_valid=df[Xv].dropna().tolist()
            y_valid=df[yv].dropna().tolist()
            self.X_train_sets.append(X_train)
            self.y_train_sets.append(y_train)
            self.X_valid_sets.append(X_valid)
            self.y_valid_sets.append(y_valid)
    def get_train_split(self,fold):
        return self.X_train_sets[fold],self.y_train_sets[fold]
    def get_valid_split(self,fold):
        return self.X_valid_sets[fold],self.y_valid_sets[fold]        
    def validate(self):
        grow_list = []
        for i in range(self.FOLDS):
            Xt,yt = self.get_train_split(i)
            Xv,yv = self.get_valid_split(i)
            print('Fold',i,'train/valid',len(Xt),len(Xv))
            if len(yt)!=len(Xt):
                print('Unequal train lengths')
            if len(yv)!=len(Xv):
                print('Unequal valid lengths')
            if 'TCGA-S9-A7J3' in Xt:
                print('TCGA-S9-A7J3 in train')
            if 'TCGA-S9-A7J3' in Xv:
                print('TCGA-S9-A7J3 in valid')
            grow_list += Xt
            grow_list += Xv
        all_cases = np.unique ( grow_list )
        for i in range(self.FOLDS):
            Xt,yt = self.get_train_split(i)
            Xv,yv = self.get_valid_split(i)
            my_cases = np.sort ( Xt+Xv )
            for X in all_cases:
                if X not in my_cases:
                    print('Fold',i,'case',X,'is missing')
            for X in my_cases:
                if X not in all_cases:
                    print('Fold',i,'case',X,'is extra')
        return True
```

```python
total = all_splits()
total.load_csv(SPLITS_PATH+SPLITS_FILE)
X,y = total.get_train_split(0)
X,y = total.get_train_split(4)
if total.validate():
    print('ok')
else:
    print('problem!')
```
