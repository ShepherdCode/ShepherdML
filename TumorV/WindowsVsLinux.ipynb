# Compare CellProfiler outputs 
Establish there are only trivial differences between our Windows and Linux runs on the same data.

```python
import numpy as np
import pandas as pd
import os
from datetime import datetime
print(datetime.now())
```

```python
WINDOWS_PATH='/home/jrm/Adjeroh/Glioma/WindowsVsLinux/TCGA-HW-7495-01Z-00-DX1.Windows/'
LINUX_PATH='/home/jrm/Adjeroh/Glioma/August_Run/CellProfilerOutputs/Output1/'
FILENAME='Process100_Image.csv'
TARGET = 'TCGA-HW-7495-01Z-00-DX1'
```

```python
windows = pd.read_csv(WINDOWS_PATH+FILENAME)
print(len(windows))
```

```python
windows = windows.loc[windows['FileName_Tumor'].str.startswith(TARGET)]
windows.describe()
```

```python
linux = pd.read_csv(LINUX_PATH+FILENAME)
print(len(linux))
```

```python
linux = linux.loc[linux['FileName_Tumor'].str.startswith(TARGET)]
linux.describe()
```

```python
for c in windows.columns:
    if c not in linux.columns:
        print(c)
```

```python
for c in linux.columns:
    if c not in windows.columns:
        print(c)
```

```python
for d in linux.dtypes:
    if (d not in ['float64','int64']):
        print(d)
```

```python
for d in windows.dtypes:
    if (d not in ['float64','int64']):
        print(d)
```

```python
for c in linux.columns:
    if (linux[c].dtype in ['float64','int64']):
        if not c.startswith('ExecutionTime_'):
            lmean=linux[c].mean()
            wmean=windows[c].mean()
            if not np.isclose(lmean,wmean,rtol=0.001):
                if not np.isnan(wmean) and not np.isnan(lmean):
                    print(c,lmean,wmean)
```
