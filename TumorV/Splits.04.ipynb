```python
import datetime
print(datetime.datetime.now())
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
import csv
import os
```

```python
GIVEN='/home/jrm/Adjeroh/Glioma/August_Run/TrainValidSplits/cvfolds.csv'
LABFILE='/home/jrm/Adjeroh/Glioma/August_Run/TrainValidSplits/aug_case_labels.csv'
OUTFILE='/home/jrm/Adjeroh/Glioma/August_Run/TrainValidSplits/aug_train_valid_splits.csv'
IDS_DIR='/home/jrm/Adjeroh/Glioma/August_Run/CellProfilerPerPatient/'
```

Get full case IDs from the directory containing one subdirectory per patient.

```python
PATIENT_NAMES = os.listdir(IDS_DIR)
PATIENT_NAMES = [name for name in PATIENT_NAMES if name.startswith('TCGA-')]
PATIENT_NAMES
```

Get the cancer class labels from the problematic train/test splits file.

```python
df = pd.read_csv(GIVEN)
df
```

For convenience, write a file that equates patient to cancer class.  
Extend the short patient name in the problematic file to the longer case IDs loaded above.

```python
def get_case_ID(partial):
    patients = [x for x in PATIENT_NAMES if x.startswith(partial)]
    if len(patients) != 1:
        raise Exception("Patient prefix not expandable to patient")
    patient = patients[0]
    #print('Convert',partial,'to',patient)
    return patient
```

```python
with open (LABFILE,'w') as outfile:
    mywriter = csv.writer(outfile)
    for idx,row in df.iterrows():
        if pd.isnull(row['f0_TRAIN_X']):
            raise Exception("null train value???")
        partial = row['f0_TRAIN_X']
        patient = get_case_ID(partial) 
        label = str(int(row['f0_TRAIN_y']))
        mywriter.writerow(( patient, label ))
    for idx,row in df.iterrows():
        # Since num_valid<num_train, file has null valid names in some rows.
        if not pd.isnull(row['f0_TEST_X']):
            partial = row['f0_TEST_X']
            patient = get_case_ID(partial) 
            label = str(int(row['f0_TEST_y']))
            mywriter.writerow(( patient, label ))
```

```python
X=[]
y=[]
with open (LABFILE,'r') as labels:
    myreader = csv.reader(labels)
    for row in myreader:
        X.append(row[0])
        y.append(row[1])
```

```python
# Reproducible stratified K-fold cross validation sets.
skf = StratifiedKFold(n_splits=5, random_state=12345, shuffle=True)
fold = 0
with open (OUTFILE,'w') as folds:
    mywriter = csv.writer(folds)
    row = ('Fold','Split','Case','Class')
    mywriter.writerow(row)
    for train_index, test_index in skf.split(X, y):
        for i in train_index:
            row = (fold,'Train',X[i],str(y[i]))
            mywriter.writerow(row)
        for i in test_index:
            row = (fold,'Valid',X[i],str(y[i]))
            mywriter.writerow(row)
        fold += 1
```
