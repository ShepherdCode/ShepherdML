## Count co-occurrence

```python
df_nc = pd.read_csv ('ncRNA.2mer.features.csv')
df_pc = pd.read_csv ('pcRNA.2mer.features.csv')

row = 0 # first sequence
col = 3 # second K-mer because 0=seqnum, 1=seqlen, 2='AA', 3=='AC'
nc_occ_3=df_nc.iloc[row][col]
nc_occ_2=df_nc.iloc[row][col-1]
(nc_occ_2,nc_occ_3)
```

```python
pc_occ_3=df_pc.iloc[row][col]
pc_occ_2=df_pc.iloc[row][col-1]
(pc_occ_2,pc_occ_3)
```

```python
df_nc.columns[2]
```

```python
num_kmers=16
num_seqs=10000 # reasonable size sample: first 10K of nc and first 10K of pc
first_kmer_column=2
co_occur=[]
for i in range(1,num_kmers):
    I = first_kmer_column+i
    kmer_i = df_nc.columns[I]
    for j in range(0,i):
        J=first_kmer_column+j
        kmer_j = df_pc.columns[J]
        NC_sum = 0
        PC_sum = 0
        for s in range(num_seqs):
            nc_ci = df_nc.iloc[s][I]
            nc_cj = df_nc.iloc[s][J]
            nc_min = min(nc_ci,nc_cj)
            nc_len = df_nc['seqlen'][s]
            NC_sum += nc_min/nc_len
            pc_ci = df_pc.iloc[s][I]
            pc_cj = df_pc.iloc[s][J]
            pc_min = min(pc_ci,pc_cj)
            pc_len = df_pc['seqlen'][s]
            PC_sum += pc_min/pc_len
        R = NC_sum/PC_sum
        print("R %s / %s = %10f"%(kmer_i,kmer_j,R))
        triple=(kmer_i,kmer_j,R)
        co_occur.append(triple)
```

```python
f=open('triples.txt','w')
for ele in co_occur:
    f.write(str(ele))
    f.write('\n')
f.close()
```

```python
df=pd.DataFrame(co_occur)
df.columns=['mer1','mer2','freq']
df['freq'].min(),df['freq'].max()
```
