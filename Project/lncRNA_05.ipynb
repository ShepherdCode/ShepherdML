Combine K-mer counts for multiple K.
Yeah, our program should generate these.

```python
import pandas as pd
df2 = pd.read_csv ('ncRNA.2mer.features.csv')
df3 = pd.read_csv ('ncRNA.3mer.features.csv')
```

```python
df2
```

```python
df3
```

```python
dfm=df3.drop(columns=['seqnum','seqlen'])
```

```python
dfc=pd.concat([df2,dfm],axis='columns')
```

```python
dfc
```

```python
dfc.to_csv('ncRNA.2-3.features.csv')
```
