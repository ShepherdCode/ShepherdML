# Principal Component Analysis
The most popular form of dimensionality reduction.

```python
from sklearn.datasets import load_iris
iris = load_iris()
# 4 features predicts species 
X = iris.data 
y = iris.target
X[:5]
```

```python
# Create 4 feature vectors
import numpy as np
f1,f2,f3,f4 = np.split(X,4,axis=1)
# Plot first three features in 3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(f1,f2,f3,c=y)
plt.show()
```

```python
from sklearn.decomposition import PCA
# How many dimension is the right number?
pca = PCA()
pca.fit(X)
explained = pca.explained_variance_ratio_
explained
```

```python
# This shows that the first 2 pcs explain 97% of variance.
# Book shows how to pick the 95% confidnce level with numpy.cumsum() and numpy.argmax().
# Here is how to ask PCA to pick enough components to explain 95% of variance.
pca = PCA(n_components=0.95)
```

```python
# Here is how to get the first 3 principal components.
pca = PCA(n_components=3)
X3d = pca.fit_transform(X)

# Break it into one vector per principal component.
pc1,pc2,pc3 = np.split(X3d,3,axis=1)
X3d[:5]
```

```python
# 3D viz, 3 principal components
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pc1,pc2,pc3,c=y)
plt.show()
```

```python
# 2D viz, 2 principal components
import matplotlib.pyplot as plt
plt.scatter(pc1,pc2,c=y)
plt.show()
```
