```python
import csv
```

```python
GUDENAS_TID_FILE = \
'/Users/jasonmiller/WVU/Localization/Gudenas/'+\
'Gudenas.Supplement.Data1.csv'
```

```python
gudenas_tids=[]
with open (GUDENAS_TID_FILE,'r') as infile:
    reader = csv.reader(infile)
    header = next(reader)
    for row in reader:
        gudenas_tids.append(row[0])
print(len(gudenas_tids))
```

```python
GENCODE_GFF_FILE = \
'/Users/jasonmiller/WVU/Localization/GenCode/'+\
'gencode.v42.lncRNA_transcripts.csv'
```

```python
tid_to_gid={}
with open (GENCODE_GFF_FILE,'r') as infile:
    header = next(infile)
    for row in infile:
        trunc = row[:100]
        fields = trunc.split(',')
        versioned = fields[0]
        tid = versioned.split('.')[0]
        gid = fields[1]
        tid_to_gid[tid]=gid
print(len(tid_to_gid.keys()))
```

```python
gid_counts = {}
known_transcripts = 0
unknown_transcripts = 0
unknown_ids = []
genes=0
dupes = 0
for tid in gudenas_tids:
    if tid not in tid_to_gid:
        unknown_transcripts += 1
        unknown_ids.append(tid)
        #print('Unknown transcript',tid)
    else:
        known_transcripts += 1
        gid = tid_to_gid[tid]
        if gid not in gid_counts:
            gid_counts[gid] = 1
            genes += 1
        else:
            gid_counts[gid] += 1
            dupes += 1
print('Unknown transcripts:',unknown_transcripts)
print('Known transcripts:',known_transcripts)
print('Genes used:',genes)
print('Duplicate transcripts:',dupes)
exactly_two = 0
more_than_two = 0
for gid in gid_counts:
    if gid_counts[gid]>=2:
        if gid_counts[gid]==2:
            exactly_two += 1
        else:
            more_than_two += 1
print('Genes with exactly two transcripts:',exactly_two)
print('Genes with more than two transcripts:',more_than_two)
```
