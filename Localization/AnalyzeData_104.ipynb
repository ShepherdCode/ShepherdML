# Missing Genes
Some genes are listed in lncAtlas but not in GenCode.

List genes in LncAtlas not found in GenCode FASTA. These are old or not mapped to chromosomes.

List genes in LncAtlas not found in GenCode GFF. These are old.

Since the two lists are the same, for coding and non-coding,
we conclude that lncAtlas does not contain any genes not mapped to chromosomes.

```python
GFF_FILE='/Users/jasonmiller/WVU/Localization/GenCode/GenCode43/gencode.v43.chr_patch_hapl_scaff.annotation.gff3'
PC_ATLAS_FILE='/Users/jasonmiller/WVU/Localization/LncAtlas/pc_CNRCI_not_null.csv'
NC_ATLAS_FILE='/Users/jasonmiller/WVU/Localization/LncAtlas/nc_CNRCI_not_null.csv'
PC_CANON_FILE='/Users/jasonmiller/WVU/Localization/GenCode/GenCode43/gencode_v43.canon_pc_transcripts.csv'
NC_CANON_FILE='/Users/jasonmiller/WVU/Localization/GenCode/GenCode43/gencode_v43.canon_lncRNA_transcripts.csv'
```

```python
GFF_GENES = set()
with open (GFF_FILE) as fin:
    for line in fin:
        if line[0] != '#':
            fields = line.split('\t')
            if fields[2] == 'gene':
                tags = fields[8]
                gene_id = tags[3:18]  # ID=ENSG00000290825
                GFF_GENES.add(gene_id)
print(len(GFF_GENES),'genes from GFF')
```

```python
def list_not_found(CANON_FILE,ATLAS_FILE):
    gencode_genes = set()
    with open (CANON_FILE) as fin:
        # csv with headers transcript_id,gene_id,biotype,length,sequence
        header = None
        for line in fin:
            line = line.strip()
            if header is None:
                header = line
            else:
                fields = line.split(',')
                gene_id=fields[1]
                gencode_genes.add(gene_id)
    gff_missing = set()
    fasta_missing = set()
    with open (ATLAS_FILE) as fin:
        # csv with headers ENSEMBL ID,Data Source,Data Type,Value,Gene Name,Coding Type,Biotype
        header = None
        for line in fin:
            line = line.strip()
            if header is None:
                header = line
            else:
                fields = line.split(',')
                gene_id=fields[0]
                if gene_id not in gencode_genes:
                    fasta_missing.add(gene_id)
                if gene_id not in GFF_GENES:
                    gff_missing.add(gene_id)
    print(len(fasta_missing),'not found in fasta')
    print(fasta_missing)
    print()
    print(len(gff_missing),'not found in gff')
    print(gff_missing)
```

```python
list_not_found(PC_CANON_FILE,PC_ATLAS_FILE)
```

```python
list_not_found(NC_CANON_FILE,NC_ATLAS_FILE)
```
