## WSI
Following Hou et al 2016, 
we selected WSI from two TCGA datasets:
Glioblastoma Multiforme (TCGA-GBM) 
and Low-grade Gioblastoma (TCGA-LGG).

Following Hou et al 2016,
we assigned WSI one of six cancer class labels based on 
diagnosis labels in the TCGA metadata.
These labels are 0 (for GBM) and 1 thru 5 (for LGG subclasses).

We collected a subset of the WSI at TCGA. 
Specifically, we used WSI IDs kindly provided by Le Hou.
These IDs formed the test set in one random train/test split,
of the 2016 published study (Le Hou, personal communication).

All of these WSI were generated by one imaging center, 
indicated by DX in the WSI ID.
Some tumor samples were imaged more than once,
indicated by numbers like DX1, DX2, ... DX9 in the WSI ID.
Some tumors could be represented by multiple samples,
and some patients could be represented by multiple tumors.
